```python
import pandas as pd
```

```python
df = pd.read_csv("Titanic.csv")
```

```python
df.head()
```

```python
#Drop some columns that need to help
df.drop(["sibsp","parch","embarked"],inplace=True,axis = 1)
```

```python
df.head()
```

```python
#Now feature encoding part in the program
dummies = pd.get_dummies(df.sex,drop_first=True)
```

```python
dummies
```

```python
df = pd.concat([dummies,df],axis = 1)
```

```python
df.drop("sex",inplace=True,axis = 1)
```

```python
df.head()
```

```python
#Label encoding
df["class"] = df["class"].map({"First": 1, "Second": 2, "Third":3})
```

```python
df.head()
```

```python
#Now One hot encoding
dummies = pd.get_dummies(df.who,drop_first=True)
```

```python
df = pd.concat([dummies,df],axis = 1)
```

```python
df.head()
```

```python
df.drop("alone",axis =1, inplace =True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df["age"]= df.age.fillna(df.age.mean())
```

```python
df.drop(["who"],axis = 1, inplace = True)
```

```python
# Now Train test split
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train
```

```python
X.columns = X.columns.astype(str)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()

classifier.fit(X_train, y_train)
```

```python
predict_data = classifier.predict(X_test)
```

```python
predict_data
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, predict_data)
print(f"Accuracy: {accuracy}")
```

```python
result = classifier.predict([[0.0,1.0,0.0,35.0,53.1,1.0]])
if result[0] == 0:
  print("Not survied")
else :
  print("Survied")
```
